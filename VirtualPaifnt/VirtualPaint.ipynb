```python
import cv2 as cv
import numpy as np
cap = cv.VideoCapture(0)

myColors = [[5,207,0,19,255,255],  # colors from trackbars
           [133,56,0,159,156,255],
           [57,76,0,100,255,255],
           [98,48,0,118,255,255]]

colorValue = [[51,153,255], #code coloring
             [255,0,255],
             [0,255,0],
             [255,0,0]]

myPoint = [] ##[ x,y,colorid ] # defining list for drawing

# Check if the webcam is opened correctly
if not cap.isOpened():
    raise IOError("Cannot open webcam")
    
def findColors(frame,myColors,colorValue):  # function for defining colors
    imgHSV = cv.cvtColor(frame,cv.COLOR_BGR2HSV)
    count = 0
    newPoint=[]
    for color in myColors:
        lower = np.array(color[0:3]) # slicing color from taskbars with defining variables
        upper = np.array(color[3:6])

        mask = cv.inRange(imgHSV,lower,upper) # calling variables with variable mask
        x,y = getContours(mask)
        
        cv.circle(imgResult,(x,y),10,colorValue[count],cv.FILLED) # drawing circle
        if x!=0 and y!=0:
            newPoint.append([x,y,count])
        count+=1

        #cv.imshow(str(color[0]),mask)
    return newPoint

def getContours(frame):
    contours,hierarchy = cv.findContours(frame,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_NONE)# drawing shape for point to draw
    x,y,w,h=0,0,0,0
    for cnt in contours:
        area = cv.contourArea(cnt)
        if area>500:
            #cv.drawContours(imgResult,cnt,-1,(255,0,0),3)   # need this function to locate exact point of drawing pencil
            peri = cv.arcLength(cnt,True)
            approx = cv.approxPolyDP(cnt,0.02*peri,True)
            x,y,w,h = cv.boundingRect(approx)
    return x+w//2,y

def drawOnCanvas(myPoint,colorValue):  # function is for drawing
    for point in myPoint:
          cv.circle(imgResult,(point[0],point[1]),10,colorValue[point[2]],cv.FILLED)
            
        
            
        

while True: # want to run code repeatly so that we can contineously draw
    ret, frame = cap.read()
    frame = cv.resize(frame, None, fx=0.5, fy=0.5, interpolation=cv.INTER_AREA)
    newPoint = findColors(frame,myColors,colorValue)
    if len(newPoint)!=0:
        for newP in newPoint: # for loop because above list cant be inside list
            myPoint.append(newP)
    if len(myPoint)!=0:
        drawOnCanvas(myPoint,colorValue)
    cv.imshow('Input', imgResult)
    imgResult = frame.copy()
    

    c = cv.waitKey(1)
    if c == 27:
        break

cap.release()
cv.destroyAllWindows()
    
```
